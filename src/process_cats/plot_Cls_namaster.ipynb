```python
clf = pk.load(open('/global/project/projectdirs/des/shivamp/ACTxDESY3_data/measurements/Cls_namaster/cls_DES_redmagic_v6.22.2_ACT_y_20191015_nside4096_linear.pk','rb' ) )



```

```python
cl_yy = clf[ ('ymap', 'ymap')][('y','y')]['true'][0]
ell = clf['ell']
plotf = ell*(ell+1)/(2*np.pi)
```

```python
fig, ax = plt.subplots(1,1, figsize = (8,6))
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(ell ,plotf* cl_yy, 'r-', label=r'$y-y$')
ax.legend(fontsize=18)
plt.xlabel(r'$\ell$', fontsize=22)
# plt.ylabel(r'$C_{\ell}$', fontsize=26)
plt.ylabel(r'$\ell (\ell+1) C_{\ell} /2\pi$', fontsize=22)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.tick_params(axis='both', which='minor', labelsize=15)
plt.tight_layout()


```

```python
bins = [1,2,3,4,5]


fig, ax = plt.subplots(1,1, figsize = (8,6))
ax.set_xscale('log')
ax.set_yscale('log')

for binv in bins:
    cl_gg = clf[ ('galaxy_density', 'galaxy_density')][(binv,binv)]['true'][0]
    ax.plot(ell ,plotf* cl_gg, label=r'Bin ' + str(binv),lw=2)
ax.legend(fontsize=18)
plt.xlabel(r'$\ell$', fontsize=22)
# plt.ylabel(r'$C_{\ell}$', fontsize=26)
plt.ylabel(r'$\ell (\ell+1) C_{\ell} /2\pi$', fontsize=22)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.tick_params(axis='both', which='minor', labelsize=15)
plt.tight_layout()


```
